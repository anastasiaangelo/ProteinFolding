# Calls to QPU small systems

```python
df_qpu = df_min_shots.loc[:, ['num_res', 'num_rot', 'num_qubits', 'shots', 'p', 'min_iter']].copy()
df_qpu['qpu_calls'] = df_qpu['min_iter'] * df_qpu['shots'] 

df_qpu.head()
```

```python
file_path = os.path.join(root_dir, "notebooks/local_testing/QPU_calls_small.csv")
file_exists = os.path.exists(file_path)

if file_exists:
    df_existing = pd.read_csv(file_path)

    df_qpu = df_qpu[~df_qpu.apply(tuple, axis=1).isin(df_existing.apply(tuple, axis=1))]

    if df_qpu.empty:
        print("No new data to append.")
    else:
        df_qpu.to_csv(file_path, mode='a', index=False, header=False)
        print("Appended new unique rows.")
else:
    df_qpu.to_csv(file_path, mode='w', index=False, header=True)
    print("Created new file and saved data.")

print(df_qpu)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.interpolate import interp1d

cpu_df = pd.read_csv("CPU_calls.csv")  
qpu_df = pd.read_csv("QPU_calls_small.csv") 

df = pd.merge(cpu_df, qpu_df, on="num_qubits", how="outer", suffixes=('_cpu', '_qpu'))

qpu_stats = df.groupby("num_qubits")["qpu_calls"].agg(["mean", "std"]).reset_index()
qpu_stats.rename(columns={"mean": "qpu_mean", "std": "qpu_std"}, inplace=True)
qpu_stats.dropna(inplace=True)

print("Unique num_qubits values:", qpu_stats["num_qubits"].values)
print("QPU Mean Values:", qpu_stats["qpu_mean"].values)

interp_func = interp1d(qpu_stats["num_qubits"], qpu_stats["qpu_mean"], kind='cubic', fill_value="extrapolate")

num_qubits_smooth = np.linspace(qpu_stats["num_qubits"].min(), qpu_stats["num_qubits"].max(), 100)
qpu_interpolated = interp_func(num_qubits_smooth)

plt.figure(figsize=(10, 6))

plt.plot(df['num_qubits'], df['cpu_calls'], marker='s', linestyle='--', label="CPU Calls", color='darkorange')

plt.errorbar(qpu_stats["num_qubits"], qpu_stats["qpu_mean"], yerr=qpu_stats["qpu_std"], fmt='o', label="QPU Calls (Mean ± Std Dev)", color='royalblue')

if len(qpu_interpolated) > 0:
    plt.plot(num_qubits_smooth, qpu_interpolated, linestyle="-", label="QPU Interpolation", color="royalblue")

plt.xlabel("Number of Qubits")
plt.ylabel("Number of Calls")
plt.legend()
plt.grid(True)

plt.savefig("cpu_vs_qpu_calls_with_errorbars.pdf")
plt.show()
plt.close()
```
